```python
import tensorflow as tf
import os
import numpy as np
import pandas as pd
from uritools import uricompose, urijoin, urisplit, uriunsplit
```

```python
curr_dir = os.getcwd()

dataFrame = pd.read_csv('data.csv')

def getAudioName(row):
    strategy = row['strategy']
    url = row['url']
    label = row['label']
    
    if (strategy == "youtube"):
        url_data = urisplit(url)
        video_id = url_data.getquerydict().get("v")[0]
        return label + "_" + video_id + ".wav"
    else:
        print("Strategy "+strategy+" not defined!")

print('Adding file name as extra columns')
dataFrame['fileName'] = dataFrame.apply (getAudioName, axis=1)

print('After adding file name: \n', dataFrame.head(1))

def preprocess(data, test_data_split):
    """Splits data to train and test set

    Parameters:
    data (DataFrame): data frame
    test_data_split (float): percentage of test data

    Returns:
    tuple: returns test and train data

    """
    if(test_data_split >= 1):
        raise Exception("Splitting test data 100% or greater is invalid!", 'test_data_split')
    else:
        rows_count = data.shape[0]
        train_data_count = int(round(rows_count * (1- test_data_split), 0))
        test_data_count = int(round(rows_count * test_data_split, 0))
        data = data.reindex(np.random.permutation(data.index))
        train, test = (data.head(train_data_count), data.tail(test_data_count))
    
    return train, test
```

```python
from python_speech_features import mfcc
import wave as wv_tool
import librosa.display
import matplotlib.pyplot as plt

PATH_TO_DATA = "information_collection/audio_data/"

def getFrameRate(filePath):
    frame_rate=0
    with wv_tool.open(filePath, "rb") as wave_file:
        frame_rate = wave_file.getframerate()
        
    if (frame_rate == 0):
        raise Exception("Frame rate is set 0! May missing files or wrong file paths!", 'frame_rate')
    
    #return frame_rate
    return 8000

def displayMfccExamples(mfcc_samples):
    sample_count = len(mfcc_samples.keys())
    fig, axes = plt.subplots(nrows=1, ncols=sample_count, sharex=False, sharey=True,figsize=(20,5))
    i=0
    for x in range(1):
        for y in range(sample_count):
            title = list(mfcc_samples.keys())[i]
            axes[y].set_title(title)
            axes[y].imshow(list(mfcc_samples.values())[i],
                            cmap='hot', interpolation='nearest')
            axes[y].get_xaxis().set_visible(False)
            axes[y].get_yaxis().set_visible(False)
            i += 1
    return

classes = dataFrame.label.unique()
samples = {}
print(classes)
for aClass in classes:
    fileName = dataFrame[dataFrame.label == aClass].iloc[0].fileName
    filePath = PATH_TO_DATA + fileName
    print (filePath)
    print(aClass)
    w, rate = librosa.load(filePath, sr=getFrameRate(filePath))
    samples[aClass] = mfcc(w[:rate], rate*1.5)

displayMfccExamples(samples)
plt.show()
```

```python
from keras.layers import Conv2D, MaxPooling2D, Dropout, Dense, Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from sklearn.utils.class_weight import compute_class_weight
from tqdm import tqdm
import random

labels_count = len(classes)

X = []

y = []

def getOneHotVectorViaClass(aClass):
    arr = np.zeros(labels_count)
    i = np.where(classes == aClass)
    arr[i] = 1
    return arr

MAX_LEN = 50

MFCC_FEA_NUM = 20

def getSmallSample(wave, samplingRate, mfcc, max_len=MAX_LEN):
    if (max_len > mfcc.shape[1]):
        print('gettin last parts')
        pad_width = max_len - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad_width)), mode='constant')
    else:
        random_pos = random.randrange(0, mfcc.shape[1] - max_len, 1)
        mfcc = mfcc[:, random_pos : random_pos + max_len]
    return mfcc

def appendXnY(aClass, wave, sr, mfcc):
    y.append(getOneHotVectorViaClass(aClass))
    audio_mfcc = getSmallSample(wave=wave, samplingRate=sr, mfcc=mfcc)
    X.append(audio_mfcc)

def initTrainData(trainData):
    for index, row in tqdm(trainData.iterrows()):
        label = row["label"]
        filePath = PATH_TO_DATA + row["fileName"]
        frameRate = getFrameRate(filePath)
        wave, sr = librosa.load(filePath, mono=True, sr=frameRate)
        print(sr)
        mfcc = librosa.feature.mfcc(wave, n_mfcc=MFCC_FEA_NUM, sr=sr)
        counter=int(int(mfcc.shape[1])/MAX_LEN)*30
        print(counter)
        for i in tqdm(range(counter)):
            print(filePath)
            appendXnY(label, wave, sr, mfcc)

initTrainData(dataFrame)

```
